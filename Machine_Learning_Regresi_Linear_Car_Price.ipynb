# Import Library

```python
#Import library pandas
import pandas as pd

#Import library numpy
import numpy as np

#Import library matplotlib dan seaborn untuk visualisasi
import matplotlib.pyplot as plt
import seaborn as sns

#Import modul LinearRefression digunakan untuk memanggil algoritma linear regression
from sklearn.linear_model import LinearRegression

#Import modul train_test_split digunakan untuk membagi data kita menjadi traning dan testing set
from sklearn.model_selection import train_test_split

#Import math agar program dapat menggunakan semua fungsi yang ada pada modul  math(ex:sqrt)
import math

#Me-non aktifkan peringatan pada python
import warnings
warnings.filterwarnings('ignore')
```

# Cek Dataset

```python
data ='CarPrice_Assignment.csv'
dataset = pd.read_csv(data)
dataset.head(10)
```

```python
dataset.info()
```

```python
dataset.isna().sum()
```

```python
dataset.corr()
```

```python
def pp(x,y,z):
    sns.pairplot(dataset,x_vars=[x,y,z], y_vars='price', size=4, aspect=1, kind='scatter')
    plt.show()

pp('enginesize','boreratio','stroke')
pp('compressionratio','horsepower','peakrpm')
pp('wheelbase','citympg','highwaympg')
```

```python
plt.scatter(dataset['enginesize'], dataset['price'])
plt.xlabel('enginesize')
plt.ylabel('price')
plt.title('Scatter Plot enginesize vs price')
plt.show()
```

# Membangun Model
 Dari hasil visualisasi diatas bahwa fitur/kolom enginesize memiliki korelasi yang tinggi terhadap kolom price/variabel dependent sehingga kita mengambil fitur/kolom enginesize untuk di traning


*   Independent variabel(x) adalah enginesize
*   Dependent variabel(y) adalah price




```python
#Mengubah bentuk matriks dengan reshape
x = dataset['enginesize'].values.reshape(-1,1)
y = dataset['price'].values.reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=100)
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
#Bagian ini cuma mengecek nilai koefisien dan intersep
#Boleh diskip
print(regressor.coef_)
print(regressor.intercept_)
```

```python
#Cek score model
regressor.score(x_test,y_test)
```

```python
#Prediksi data test
y_prediksi =regressor.predict(x_test)

plt.scatter(x_test,y_test)
plt.plot(x_test,y_prediksi, c='r')
plt.xlabel('enginesize')
plt.ylabel('price')
plt.title('plot enginesize vs price')
```

```python
print('nilai prediksi harga dengan enginesize 100 :',regressor.predict([[90]]))
print('nilai prediksi harga dengan enginesize 150 :',regressor.predict([[150]]))
print('nilai prediksi harga dengan enginesize 200 :',regressor.predict([[200]]))
```
